```scala
%pyspark

def parse_movie_record(s):
    id = s.split(",")[0]
    genres = s.split(",")[-1]
    name = ','.join(s.split(",")[1:-1])
    name = name.replace('"','')
    return (id,name,genres)

movies = sc.textFile('gs://oklev-uku-datasets/movie-ratings/movies.csv')
movies = movies.map(parse_movie_record)

print(movies.take(5))

ratings = sc.textFile('gs://oklev-uku-datasets/movie-ratings/ratings.csv')

print(ratings.take(5))
```

For each user, find total number of ratings provided.

```scala
%pyspark
user_rating_counts = (ratings
    .filter(lambda line: not line.startswith("userId"))  
    .map(lambda line: line.split(",")[0])               
    .map(lambda userId: (userId, 1))
    .reduceByKey(lambda a, b: a + b)                     
)

user_rating_counts.collect()
```

How many movies an average user has ranked?

```scala
%pyspark
user_rating_counts.map(lambda x: x[1]).mean()
```

For each user, find the minimum, maximum and average rating he provided


```scala
%pyspark
user_stats = (ratings
    .filter(lambda line: not line.startswith("userId"))
    .map(lambda line: line.split(","))
    .map(lambda f: (f[0], float(f[2])))
    .aggregateByKey(
        (float("inf"), float("-inf"), 0.0, 0),                
        lambda acc, r: (min(acc[0], r), max(acc[1], r), acc[2] + r, acc[3] + 1),
        lambda a, b: (min(a[0], b[0]), max(a[1], b[1]), a[2] + b[2], a[3] + b[3]),
    )
    .mapValues(lambda acc: (acc[0], acc[1], acc[2] / acc[3])) 
)
user_stats.collect()
```

Find top 10 movies by average rating


```scala
%pyspark
movies = (movies
    .filter(lambda x: x[0] != "movieId")
    .map(lambda x: (x[0], x[1]))                     
)

movie_avg = (ratings
    .filter(lambda line: not line.startswith("userId"))
    .map(lambda line: line.split(","))
    .map(lambda f: (f[1], (float(f[2]), 1)))          
    .reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1]))  
    .mapValues(lambda sc: (sc[0] / sc[1], sc[1])))      

top10 = (movie_avg
    .join(movies)                                      
    .map(lambda kv: (kv[1][1], kv[1][0][0], kv[1][0][1]))  
    .takeOrdered(10, key=lambda t: -t[1])              
)

print(top10)
```
