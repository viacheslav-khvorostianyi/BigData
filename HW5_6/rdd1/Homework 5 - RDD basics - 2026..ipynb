```scala
%pyspark
def parse_movie_record(s):
    id = s.split(",")[0]
    genres = s.split(",")[-1]
    name = ','.join(s.split(",")[1:-1])
    name = name.replace('"','')
    return (id,name,genres)

movies = sc.textFile('gs://oklev-uku-datasets/movie-ratings/movies.csv')
movies = movies.map(parse_movie_record)

movies.collect()
```

Find number of movies in the dataset


```scala
%pyspark
header = movies.first()

# 3. Filter out the header from the data RDD
data = movies.filter(lambda row: row != header)
data.count()
```

How many movies of "Comedy" genre database contains?

```scala
%pyspark
data.filter(lambda x:'Comedy' in x[-1]).count()
```

What are the yealiest and the latest years represented in the database?


```scala
%pyspark
import re


year_re = re.compile(r"\((\d{4})\)\s*$")

years = (data
    .map(lambda x: year_re.search(x[1]) if x[1] is not None else None)
    .map(lambda m: int(m.group(1)) if m else None)
    .filter(lambda y: y is not None))
print(years.max())
print(years.min())
```

What is the minimum, average and maximum number of genres per single movie?


```scala
%pyspark
genres_parsed_count = data.map(lambda x: len(x[-1].split("|")))
print("min", genres_parsed_count.min())
print("avg", genres_parsed_count.mean())
print("max", genres_parsed_count.max())

```

Show the list of reused movie names

```scala
%pyspark
reused_titles = (data
    .map(lambda x: (x[1], 1))                   
    .reduceByKey(lambda a, b: a + b)            
    .filter(lambda kv: kv[1] > 1)              
)

reused_titles.collect()
```
